```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
import os
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator

class EnhancedIDMatcher:
    def __init__(self):
        print("Initializing ID Matcher...")
        self.face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        self.siamese_model = None
        self.base_model = None
        self.data_augmentor = self.create_augmentor()
        self.threshold = 0.5

    def create_augmentor(self):
        return ImageDataGenerator(
            rotation_range=15,
            width_shift_range=0.2,
            height_shift_range=0.2,
            brightness_range=[0.6, 1.4],
            zoom_range=0.2,
            fill_mode='nearest'
        )


    def extract_face(self, image_path):
        img = cv2.imread(image_path)
        if img is None:
            return None
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = self.face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
        if len(faces) == 0:
            return None
        x, y, w, h = faces[0]
        face = img[y:y + h, x:x + w]
        return cv2.resize(face, (64, 64))

    def augment_image(self, image, n=3):
        aug_images = []
        reshaped = image.reshape((1,) + image.shape)
        aug_iter = self.data_augmentor.flow(reshaped, batch_size=1)
        for _ in range(n):
            aug_images.append(next(aug_iter)[0].astype(np.uint8))
        return aug_images

    def load_data(self, base_folder, augment=True):
        print("Loading dataset...")
        faces, labels = [], []
        person_folders = sorted([f for f in os.listdir(base_folder) if f.startswith('person')])
        for pid, folder in enumerate(person_folders):
            print(pid,folder)
            person_path = os.path.join(base_folder, folder)
            for file in os.listdir(person_path):
                if file.lower().endswith(('jpg', 'jpeg', 'png')):
                    face = self.extract_face(os.path.join(person_path, file))
                    if face is not None:
                        faces.append(face)
                        labels.append(pid)
                        if augment:
                            faces.extend(self.augment_image(face))
                            labels.extend([pid]*3)
        print(f"Total faces loaded: {len(faces)}")
        return np.array(faces), np.array(labels)

    def create_pairs(self, faces, labels):
        pairs1, pairs2, y = [], [], []
        label_dict = {label: np.where(labels == label)[0] for label in np.unique(labels)}
        for label, idxs in label_dict.items():
            for i in range(min(len(idxs), 5)):
                for j in range(i + 1, len(idxs)):
                    pairs1.append(faces[idxs[i]])
                    pairs2.append(faces[idxs[j]])
                    y.append(1)
        for _ in range(len(y)):
            a, b = random.sample(range(len(faces)), 2)
            if labels[a] != labels[b]:
                pairs1.append(faces[a])
                pairs2.append(faces[b])
                y.append(0)
        return np.array(pairs1), np.array(pairs2), np.array(y)

    def create_base_model(self):
        return models.Sequential([
            layers.Conv2D(32, 3, activation='relu', input_shape=(64, 64, 3)),
            layers.MaxPooling2D(),
            layers.Conv2D(64, 3, activation='relu'),
            layers.MaxPooling2D(),
            layers.Conv2D(128, 3, activation='relu'),
            layers.GlobalAveragePooling2D(),
            layers.Dense(128),
            layers.Lambda(lambda x: tf.math.l2_normalize(x, axis=1))
        ])

    def build_siamese(self):
        input1 = layers.Input((64, 64, 3))
        input2 = layers.Input((64, 64, 3))
        self.base_model = self.create_base_model()
        out1 = self.base_model(input1)
        out2 = self.base_model(input2)
        l1_dist = layers.Lambda(lambda tensors: tf.abs(tensors[0] - tensors[1]))([out1, out2])
        output = layers.Dense(1, activation='sigmoid')(l1_dist)
        model = models.Model([input1, input2], output)
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        self.siamese_model = model

    def save_models(self):
        self.siamese_model.save("best_siam_model.h5")
        self.base_model.save("basey_model.h5")
        print("Models saved.")

    def train(self, folder, epochs=30, batch_size=16):
        faces, labels = self.load_data(folder)
        p1, p2, y = self.create_pairs(faces, labels)
        p1, p2 = p1 / 255.0, p2 / 255.0
        X1_train, X1_test, X2_train, X2_test, y_train, y_test = train_test_split(p1, p2, y, test_size=0.2, stratify=y)
        self.build_siamese()
        print("Training model...")
        self.siamese_model.fit([X1_train, X2_train], y_train, validation_data=([X1_test, X2_test], y_test),
                               batch_size=batch_size, epochs=epochs, verbose=1)
        y_scores = self.siamese_model.predict([X1_test, X2_test]).flatten()
        fpr, tpr, thresholds = roc_curve(y_test, y_scores)
        best_index = np.argmax(tpr - fpr)
        self.threshold = thresholds[best_index]
        print(f"Optimal threshold: {self.threshold:.3f}")
        self.save_models()

    def get_embedding(self, face):
        face = face / 255.0
        face = np.expand_dims(face, axis=0)
        return self.base_model.predict(face, verbose=0)[0]

    def compare_faces(self, path1, path2):
        f1, f2 = self.extract_face(path1), self.extract_face(path2)
        if f1 is None or f2 is None:
            return False, 0.0, "Face not found"
        f1, f2 = f1 / 255.0, f2 / 255.0
        sim = self.siamese_model.predict([np.expand_dims(f1, 0), np.expand_dims(f2, 0)], verbose=0)[0][0]
        match = sim > self.threshold
        result = f"{'MATCH' if match else 'NO MATCH'} (Confidence: {sim:.2f})"
        print(result)
        return match, sim, result

def main():
    matcher = EnhancedIDMatcher()
    matcher.train("Downloads/id_detection")
    result = matcher.compare_faces("idC.jpeg", "selfie.jpeg")
    print("Final result:", result)

if __name__ == '__main__':
    main()
```

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import models, layers
import os

model = None
base_model = None
USE_EMBEDDING = False

print("Checking for model files...")
if os.path.exists("enhanced_siamese_model.h5"):
    print("Siamese model file found")
else:
    print("enhanced_siamese_model.h5 not found")

if os.path.exists("enhanced_base_model.h5"):
    print("Base model file found")
else:
    print("enhanced_base_model.h5 not found")

# Block 2: Recreate Model Architecture (Fixed)
def create_base_model_architecture():
    """Recreate the exact base model architecture"""
    model = models.Sequential([
        layers.Conv2D(32, 3, activation='relu', input_shape=(64, 64, 3)),
        layers.MaxPooling2D(),
        layers.Conv2D(64, 3, activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(128, 3, activation='relu'),
        layers.GlobalAveragePooling2D(),
        layers.Dense(128),
        layers.Lambda(lambda x: tf.math.l2_normalize(x, axis=1)),
        layers.Dense(64, activation='relu', name='embedding_output')
    ])
    return model

def create_siamese_architecture():
    """Recreate the exact Siamese model architecture"""
    input1 = layers.Input((64, 64, 3))
    input2 = layers.Input((64, 64, 3))
    base_net = create_base_model_architecture()

    out1 = base_net(input1)
    out2 = base_net(input2)

    l1_dist = layers.Lambda(lambda tensors: tf.abs(tensors[0] - tensors[1]))([out1, out2])
    output = layers.Dense(1, activation='sigmoid')(l1_dist)
    
    model = models.Model([input1, input2], output)
    return model, base_net

def align_face(self, image):
        return image

# Block 3: Load Models (Fixed)
def load_models():
    global model, base_model, USE_EMBEDDING
    try:
        print("Trying to recreate model architecture...")
        
        # Create new models with correct architecture
        siamese_model, base_net = create_siamese_architecture()
        
        # Try to load weights
        try:
            siamese_model.load_weights("enhanced_siamese_model.h5")
            model = siamese_model
            base_model = base_net
            print("Siamese model weights loaded successfully!")
            USE_EMBEDDING = True
            return True
        except Exception as weight_error:
            print(f"Weight loading failed: {str(weight_error)[:100]}...")
        
        # Try loading base model separately
        try:
            base_net.load_weights("enhanced_base_model.h5")
            base_model = base_net
            USE_EMBEDDING = True
            print("Base model weights loaded successfully!")
            return True
        except Exception as base_error:
            print(f"Base model loading failed: {str(base_error)[:100]}...")
            
    except Exception as e:
        print(f"Method 2 failed: {str(e)[:100]}...")

    try:
        print("Trying manual model reconstruction...")
 
        import h5py
        
        if os.path.exists("enhanced_base_model.h5"):
            base_net = create_base_model_architecture()
            # Try partial weight loading
            try:
                base_net.load_weights("enhanced_base_model.h5", by_name=True, skip_mismatch=True)
                base_model = base_net
                USE_EMBEDDING = True
                print("Base model loaded with skip_mismatch!")
                return True
            except Exception as e:
                print(f"Skip mismatch failed: {str(e)[:100]}...")
        
    except Exception as e:
        print(f"Method 3 failed: {str(e)[:100]}...")
    
    return False

success = load_models()
if not success:
    print("ERROR: No models could be loaded!")
    print("Suggestion: Please retrain the models or check model files.")

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')


def extract_face(image_path):
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return None
        
    img = cv2.imread(image_path)
    if img is None:
        print(f"Could not read image: {image_path}")
        return None
        
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)
    
    if len(faces) == 0:
        print(f"No face detected in: {image_path}")
        return None
        
    x, y, w, h = faces[0]
    face = img[y:y+h, x:x+w]
    return cv2.resize(face, (64, 64)) / 255.0

def compare_faces(img1_path, img2_path, threshold=0.5):
    global model, base_model, USE_EMBEDDING
    
    if model is None and base_model is None:
        return "No model loaded", 0.0
    
    face1 = extract_face(img1_path)
    face2 = extract_face(img2_path)
    
    if face1 is None or face2 is None:
        return "Face not found", 0.0
    
    face1 = np.expand_dims(face1, 0)
    face2 = np.expand_dims(face2, 0)
    
    try:
        if USE_EMBEDDING and base_model is not None:
            print("Using embedding comparison...")
            emb1 = base_model.predict(face1, verbose=0)[0]
            emb2 = base_model.predict(face2, verbose=0)[0]
            
            # Cosine similarity
            score = np.dot(emb1, emb2) / (np.linalg.norm(emb1) * np.linalg.norm(emb2))
            result = "MATCH" if score > threshold else "NO MATCH"
            
        elif model is not None:
            print("Using Siamese model...")
            score = model.predict([face1, face2], verbose=0)[0][0]
            result = "MATCH" if score > threshold else "NO MATCH"
        else:
            return "No working model available", 0.0
        
        print(f"{result} (Score: {score:.3f})")
        return result, score
        
    except Exception as e:
        print(f"Prediction error: {str(e)}")
        return "Prediction failed", 0.0


def debug_model_structure():
    if os.path.exists("enhanced_siamese_model.h5"):
        try:
            import h5py
            with h5py.File("enhanced_siamese_model.h5", 'r') as f:
                print("\nSiamese model structure:")
                def print_structure(name, obj):
                    print(f"  {name}: {type(obj)}")
                f.visititems(print_structure)
        except Exception as e:
            print(f"Could not read Siamese model structure: {e}")
    
    if os.path.exists("enhanced_base_model.h5"):
        try:
            import h5py
            with h5py.File("enhanced_base_model.h5", 'r') as f:
                print("\nBase model structure:")
                def print_structure(name, obj):
                    print(f"  {name}: {type(obj)}")
                f.visititems(print_structure)
        except Exception as e:
            print(f"Could not read base model structure: {e}")

if success:
    print("\n" + "="*40)
    print("Testing face comparison...")
    print("="*40)
    
    
    img1 = "sirID.jpg"
    img2 = "sir.jpg"
    
    print(f"Comparing: {img1} vs {img2}")
    result = compare_faces(img1, img2)
    print(f"Final Result: {result}")
else:
    print("\nCannot run comparison - no models loaded!")
```
