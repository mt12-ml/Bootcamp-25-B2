```python

import cv2
import numpy as np
image = cv2.imread('img_1.jpg')
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv_image, lower_blue, upper_blue)
new_color_bgr = np.array([0, 0, 255]) 
new_color_image = np.full_like(image, new_color_bgr)
masked_new_color = cv2.bitwise_and(new_color_image, new_color_image, mask=mask)
inverted_mask = cv2.bitwise_not(mask)
masked_original = cv2.bitwise_and(image, image, mask=inverted_mask)
final_image = cv2.add(masked_new_color, masked_original)
cv2.imshow('Original Image', image)
cv2.imshow('Mask', mask)
cv2.imshow('Final Image', final_image)
cv2.waitKey(1000000)
cv2.destroyAllWindows()
```
