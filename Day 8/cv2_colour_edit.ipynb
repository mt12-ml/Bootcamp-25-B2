```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image
img_bgr = cv2.imread(r"C:\Users\hp pro\Desktop\ML1 bootcamp\img_1 - Copy.jpg")

# Step 2: Convert to HSV
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Step 3: Define lower and upper bounds for pink color in HSV
lower_pink = np.array([140, 50, 50])
upper_pink = np.array([170, 255, 255])

# Step 4: Create a binary mask where pink areas are white
mask = cv2.inRange(img_hsv, lower_pink, upper_pink)

# Step 5: Apply mask to original image (optional)
result = cv2.bitwise_and(img_bgr, img_bgr, mask=mask)

# Convert BGR to RGB for matplotlib display
result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

# Step 6: Display
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title("Pink Regions Mask")
plt.imshow(mask, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Original Image with Pink Highlighted")
plt.imshow(result_rgb)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\hp pro\Desktop\ML1 bootcamp\img_1 - Copy.jpg")
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)


lower_pink = np.array([160, 100, 100])
upper_pink = np.array([180, 255, 255])

mask = cv2.inRange(img_hsv, lower_pink, upper_pink)

white_pixels = np.full_like(img, 255)


img_result = cv2.bitwise_and(img, img, mask=cv2.bitwise_not(mask))  # keep non-pink
pink_to_white = cv2.bitwise_and(white_pixels, white_pixels, mask=mask)  # white for pink
final_img = cv2.add(img_result, pink_to_white)  # combine

# Convert to RGB and display
final_rgb = cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Pink Converted to White")
plt.imshow(final_rgb)
plt.axis('off')
plt.tight_layout()
plt.show()
```
