#### 29. How to round away from zero a float array ? (★☆☆)

````python
#How to round away from zero a float array 
def round_away_from_zero(arr):
    return np.where(arr > 0, np.ceil(arr), np.floor(arr))
arr = np.array([1.2, -1.2, 1.7, -1.7, 0.0])
rounded_arr = round_away_from_zero(arr)
print(rounded_arr)
````

#### 41. How to sum a small array faster than np.sum? (★★☆)

````python
#### 41. How to sum a small array faster than np.sum? (★★☆)
def sum_array(arr):
    total = 0
    for num in arr:
        total += num
    return total
sum_result = sum_array(np.array([1, 2, 3, 4, 5]))
print("Sum of the array:", sum_result)
````

#### 46. Create a structured array with `x` and `y` coordinates covering the [0,1]x[0,1] area (★★☆)

#### 56. Generate a generic 2D Gaussian-like array (★★☆)

````python
import numpy as np
import matplotlib.pyplot as plt

# Size of the array (e.g., 2D grid: 100x100)
n = 100

# Create a grid of (x,y) coordinates centered at 0
x = np.linspace(-1, 1, n)
y = np.linspace(-1, 1, n)
X, Y = np.meshgrid(x, y)

# Compute 2D Gaussian
sigma = 0.4  # standard deviation
gaussian = np.exp(-(X**2 + Y**2) / (2 * sigma**2))
print(gaussian)
````
