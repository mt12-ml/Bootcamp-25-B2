```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("img_1.jpg")
print(img)
```

```python
cv2.imshow("My image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_copy = img_rgb.copy()
plt.imshow(img_rgb)
plt.title("Input Image\n")
plt.show()
```

```python
hsvFrame = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV_img", hsvFrame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
red_lower1 = np.array([0,100,100], np.uint8)
red_upper1 = np.array([10,255,255], np.uint8)
```

```python
red_lower2 = np.array([160,100,100], np.uint8)
red_upper2 = np.array([180,255,255], np.uint8)
```

```python
red_mask1 = cv2.inRange(hsvFrame, red_lower1, red_upper1)
red_mask2 = cv2.inRange(hsvFrame, red_lower2, red_upper2)
```

```python
red_mask = np.bitwise_or(red_mask1, red_mask2)
cv2.imshow("Mask", red_mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img[red_mask>0] = (255,255,255)
cv2.imshow("Red Masked", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
